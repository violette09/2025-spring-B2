```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('new_insurance_data.csv')

# Draw histograms and histogram variants for each feature or column
for column in df.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram and KDE for {column}')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.violinplot(x=df[column])
    plt.title(f'Violin Plot for {column}')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.swarmplot(x=df[column])
    plt.title(f'Swarm Plot for {column}')
    plt.show()

# Draw grouped histograms
categorical_columns = df.select_dtypes(include=['object']).columns
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data=df, x='claim_amount', hue=column, multiple="dodge", shrink=.8)
    plt.title(f'Grouped Histogram for claim_amount by {column}')
    plt.show()

# Draw a bar plot to indicate total counts of each categorical variable in a given column
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=df[column])
    plt.title(f'Bar Plot for {column}')
    plt.show()

# Find means, medians, and modes
means = df.mean()
medians = df.median()
modes = df.mode().iloc[0]

print("Means:\n", means)
print("Medians:\n", medians)
print("Modes:\n", modes)
```

```python
conclusions = """
Data Usability: The data is complete with no missing values and is likely to be usable.
Overall Shape: The distribution of numerical features like claim amount is highly skewed right.
Outliers: There are significant outliers present in the claim amount column.
Grouped Histograms: Smokers have higher claim amounts on average compared to non-smokers.
Range: The minimum and maximum values for claim amount indicate high variability.
Bin Sizes: Changing bin sizes reveals different patterns; smaller bins provide more detail.
"""

print(conclusions)
```

# 4. Storytelling With Data graph

Reproduce any graph of your choice in p. 52-68 of the Storytelling With Data book as best you can.  (The second half of chapter two).  You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the chart
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
issues = [1, 1, 2.5, 3, 3.5, 4]

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Create the 3D bar chart
xpos = np.arange(len(months))
ypos = np.zeros(len(months))
zpos = np.zeros(len(months))
dx = np.ones(len(months))
dy = np.ones(len(months))
dz = issues

ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='skyblue', edgecolor='black')

# Add labels and title
ax.set_xticks(xpos)
ax.set_xticklabels(months)
ax.set_xlabel('Months')
ax.set_ylabel('Y')
ax.set_zlabel('Number of Issues')
ax.set_title('3D Column Chart')

plt.show()
```
